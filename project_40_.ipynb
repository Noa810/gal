```python
def keep_first_four_digits_if_more_than_four(value):
  """
  מחזיר את ארבע הספרות הראשונות אם הערך מכיל יותר מארבע ספרות.
  אחרת, מחזיר את הערך המקורי כפי שהוא.

  Args:
    value: הערך לבדיקה.

  Returns:
    ארבע הספרות הראשונות אם יש יותר מארבע ספרות, אחרת הערך המקורי.
  """
  if isinstance(value, (int, float, str)):
    # נמיר את הערך למחרוזת ונחפש רצף של ספרות
    text = str(value)
    digits = re.findall(r'\d+', text)

    if digits:
      all_digits = ''.join(digits)
      if len(all_digits) > 4:
        return int(all_digits[:4]) # נחזיר כמספר שלם
      else:
        # אם יש 4 ספרות או פחות, נחזיר את המספר שהומר
        try:
          return int(all_digits)
        except ValueError:
           return value # אם לא הצלחנו להמיר למספר, נחזיר את הערך המקורי
    else:
        return value # אם לא נמצאו ספרות, נחזיר את הערך המקורי
  else:
    return value # אם זה לא מספר או מחרוזת, נחזיר את הערך המקורי
```
