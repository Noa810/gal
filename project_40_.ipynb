```python
df3_clean['RATING'] = df3_clean['RATING'].astype('float64')
df3_clean['VOTES'] = df3_clean['VOTES'].astype('float64')
```

```python
X = df3_clean.drop(['MOVIES',  'YEAR' , 'ONE-LINE' , 'Gross','STARS','GENRE'], axis=1).to_numpy()
X.shape
```

```python

y=df3_clean['VOTES'].to_numpy()
y.shape
```

```python

X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,\
                                               shuffle=True,random_state=100)

print("train: ",X_train.shape,y_train.shape)

print("test: ",X_test.shape,y_test.shape)
```

```python
model = LinearRegression()
model.fit(X_train, y_train)

y_pred = model.predict(X_test)
```

```python
mse = mean_squared_error(y_true=y_test, y_pred=y_pred)
print(f"Mean Squared Error: {mse}")

rmse = np.sqrt(mse)
print(f"Root Mean Squared Error: {rmse}")
```

```python
pipe = Pipeline([
        ('scale', StandardScaler()),
        ('knn', KNeighborsRegressor())
        ])
```

```python
param_grid = {
    'knn__n_neighbors' :[k for k in range(1,750,20)]
}
```

```python
knn_reg_1 = GridSearchCV(pipe, param_grid=param_grid, cv=5, verbose=1)
knn_reg_1.fit(X_train, y_train)
```
