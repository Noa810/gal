```python
df['Suggested to Friends/Family (Y/N %)'] = df['Suggested to Friends/Family (Y/N %)'].str.replace('%', '').str.replace('Y', '')
df['Suggested to Friends/Family (Y/N %)'] = pd.to_numeric(df['Suggested to Friends/Family (Y/N %)'], errors='coerce').astype('double')
plt.figure(figsize=(10,10))
sns.scatterplot(x = df['Average Rating'] , y = df['Suggested to Friends/Family (Y/N %)'])
plt.xlabel('Average Rating')
plt.ylabel("Suggested to Friends/Family (Y/N %)")
```


גרף זה מציג ת התפלגות הדירוג (1-10)ביחס להמלצה לבני משפחה או חברים באחוזים

# **:מסקנה**

- אפשר לראות ריכוז בזאנרים כגון דרמה,קומדיה,רומנטיקה, אימה,מדע דמיוני


```python
plt.figure(figsize=(10,10))
mean_y_by_x = df.groupby('Genre')['Suggested to Friends/Family (Y/N %)'].mean()
ordered_categories = mean_y_by_x.sort_values().index

sns.barplot(x='Genre', y='Suggested to Friends/Family (Y/N %)', data=df, order=ordered_categories)
plt.xlabel('Genre')
plt.ylabel("Suggested to Friends/Family (Y/N %)")


plt.xticks(ticks=ordered_categories, labels=ordered_categories, rotation=45, ha='right')

ax = plt.gca()
ax.plot(ordered_categories, mean_y_by_x.loc[ordered_categories].values, color='red', marker='o', linestyle='--')
```

הגרף מציג את ממוצע ההמלצות באחוזים של כל זאנר
